```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
SPI_df = pd.read_csv('SPI_index_labelled.csv',skiprows = [1]);
pd.set_option('display.expand_frame_repr', False);
pd.set_option('display.max_columns', None);

print(SPI_df.head());   
```

```python
print(SPI_df.info());
```

```python
SPI_df.date = SPI_df.date.astype(str);
print(SPI_df.columns);
```

```python
print(SPI_df.isnull().sum());
```

```python
print(SPI_df.describe());
```

```python
columnNames = list(SPI_df.columns);
for column in columnNames[3:-4]:
    SPI_df[column] = pd.to_numeric(SPI_df[column]);

print(SPI_df.dtypes);
```

```python
SPI_df.rename(columns = {'date': 'year'}, inplace = True);
SPI_df.population.fillna(23410000 , inplace = True);
SPI_df.income = SPI_df.income.replace('Not classified', 'Low income');
print(SPI_df.head());
```

```python
fig = plt.figure(figsize = (20,20));
ax = fig.gca();
SPI_df.hist(ax = ax);
fig.tight_layout();
plt.show();
plt.close();
```

```python
plt.figure(figsize = (8,5));
sns.boxplot(x = 'region', y = 'SPI.INDEX', data = SPI_df);
plt.xticks(rotation = 30);
plt.xlabel("Regions");
plt.ylabel("SPI Index total Score");
plt.show();
plt.close();
```

```python
SPI_df_Region_mean = SPI_df.groupby('region').mean().reset_index().sort_values(by = 'SPI.INDEX',ascending = False);
SPI_df_unique_region = list(SPI_df.region.unique());
print(SPI_df_Region_mean);
```

```python
plt.figure(figsize = (10,8));
ax = sns.barplot(x = 'region', y = 'SPI.INDEX', data =  SPI_df_Region_mean);
ax.bar_label(ax.containers[0]);
plt.xlabel("Regions");
plt.ylabel("SPI Index average Score");
plt.xticks(rotation = 30);
plt.title("SPI INDEX average score by region");
plt.show();
plt.close();
```

```python
SPI_df_country_mean = SPI_df[SPI_df['SPI.INDEX'] > 0].groupby(['country','region']).mean().reset_index();
print(SPI_df_country_mean.head());
```

```python
region_quartile = np.quantile(SPI_df_country_mean['SPI.INDEX'], [0.25,0.5,0.75]);
# print(region_quartile);
quartile_labels = ['1st Quartile','2nd Quartile','3rd Quartile','4th Quartile'];
# function to set specific color for datapoints in each quartile in the plot
def colorByQuartile(aList):
    colors=[];
    for item in aList:
        if item <= region_quartile[0]:
            colors.append('orange');
        elif item >= region_quartile[0] and item <= region_quartile[1]:
            colors.append('green');
        elif item >= region_quartile[1] and item <= region_quartile[2]:
            colors.append('blue');
        else:
            colors.append('black');
    return colors;

# def label_dataPoint(x, y, val, ax):
#     a = pd.concat({'x': x, 'y': y, 'val': val}, axis = 1);
#     for dPoint, point in a.iterrows():
#         ax.text(point['x'], point['y'], str(point['val']));
```

```python
# scatterplot of SPI INDEX Total score vs. region
plt.figure(figsize = (10,8));
ax = fig.add_subplot();
sns.scatterplot(x = SPI_df_country_mean['SPI.INDEX'],
                y = pd.Categorical(SPI_df_country_mean.region, SPI_df_unique_region),
                c = colorByQuartile(SPI_df_country_mean['SPI.INDEX']),
                );
plt.xlabel('Average SPI INDEX score');
plt.ylabel('Regions');
for quantile in region_quartile:
    plt.axvline(x = quantile, color = 'red');
#label_dataPoint(SPI_df_country_mean['SPI.INDEX'], SPI_df_country_mean.region, SPI_df_country_mean.country, plt.gca());
plt.title("Distribution of SPI INDEX score by countries");
plt.show();
plt.close();
```

```python
SPI_PillarsOnly_df = SPI_df;
SPI_PillarsOnly_df = SPI_PillarsOnly_df.filter(['SPI.INDEX.PIL1','SPI.INDEX.PIL2','SPI.INDEX.PIL3','SPI.INDEX.PIL4','SPI.INDEX.PIL5','SPI.INDEX']);
print(SPI_PillarsOnly_df.corr());
```

```python
pillars_descriptiveNames = ['P1.Data.Use','P2.Data.Services','P3.Data.Products','P4.Data.Sources','P5.data.Infrastructure','Data.index.score']
# heat matrix that shows correlation between the 5 pillars indexes
fig, ax = plt.subplots(figsize = (10,8));
ax = sns.heatmap(data = SPI_PillarsOnly_df.corr(),annot = True);
ax.set_xticklabels(pillars_descriptiveNames,rotation = 0, fontsize = 9);
ax.set_yticklabels(pillars_descriptiveNames, fontsize = 8.5);
fig.tight_layout()
plt.show();
plt.clf();
```

```python
SPI_per_income_df = SPI_df;
SPI_per_income_df = SPI_per_income_df.groupby(['income']).mean().reset_index().sort_values(by = 'SPI.INDEX', ascending = False);
print(SPI_per_income_df.head());
```

```python
fig, ax = plt.subplots();
ax = sns.barplot(x = 'SPI.INDEX', y = 'income', data = SPI_per_income_df);
ax.bar_label(ax.containers[0], fontsize = 8 , fmt = '%.2d');
plt.xlabel("Income Category");
plt.ylabel("");
plt.title("SPI INDEX overall score by income segment");
plt.show();
plt.close();
```

```python
pillars_names_df = ['SPI.INDEX.PIL1','SPI.INDEX.PIL2','SPI.INDEX.PIL3','SPI.INDEX.PIL4','SPI.INDEX.PIL5'];
labels = ['High','Upper middle','Lower middle','Lower'];
N_plot = 0;
fig = plt.figure(figsize = (8,9));
for i in range(len(pillars_names_df)):
    N_plot += 1;
    # subplot to create grid of plots
    ax = plt.subplot(3, 2, N_plot);
    sns.barplot(x = 'income', y = pillars_names_df[i], data = SPI_per_income_df);
    ax.bar_label(ax.containers[0], fontsize = 8 , fmt = '%.2d');
    ax.set_xticks(range(len(labels)));
    plt.xticks(fontsize = 8);
    ax.set_xticklabels(labels);
    plt.ylabel('');
    plt.xlabel('');
    plt.title(f'{pillars_descriptiveNames[i]} Index');
plt.tight_layout();
plt.show();
         
```

```python
SPI_df['Over 70'] = SPI_df['SPI.INDEX'] >= 70;
SPI_df['Below 70'] = SPI_df['SPI.INDEX'] < 70;
SPI_Index_70_df = SPI_df;
SPI_Index_70_df = SPI_Index_70_df.filter(['income','Over 70','Below 70']);
SPI_Index_70_df = SPI_Index_70_df.groupby('income').sum().reset_index();
SPI_Index_70_df = SPI_Index_70_df.assign(Percentage = lambda x: (x['Below 70'] / (x['Below 70'] + x['Over 70']) * 100));
print(SPI_Index_70_df);
```
